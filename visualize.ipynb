## Training with images (20% training dataset = 16% entire dataset)

```python
model_dict, model_img = load_model('NN_e25_lr5e-04_small0.pt')
```

```python
def plot_loss(model):
    plot(
        range(1, model['total_epoch']+1), 
        model['train_loss'],
        'Training Loss',
        'Loss',
        color='midnightblue'
    )
    plot(
        range(1, model['total_epoch']+1), 
        model['test_loss'],
        'Validation Loss',
        'Loss'
    )
    plt.legend()
    plt.show()
    plot(
        range(1, model['total_epoch']+1), 
        model['metrics']['kappa'],
        'Validation Quadratic Weighted Kappa',
        'Kappa'
    )
    plt.legend()
    plt.show()
```

```python
plot_loss(model_dict)
```

### Load Data

```python
image_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]
    )
])
data_train = PetDataset(data_type='train', shuffle=False, small=False, image_transform=image_transform, oe=model_dict['cat_enc'])
data_test = PetDataset(data_type='test', indicies=data_train.indicies, image_transform=image_transform, oe=model_dict['cat_enc'])
```

### Categorical Embedding of Data

```python
emb2d = plot_categorical_embedding(model_img, data_test)
emb3d = plot_categorical_embedding(model_img, data_test, dim=3)
```

```python
data_test.X['cat'].columns # list of categorical features
```

#### Plot categorical embedding with t-SNE

```python
# get original categorical encoding value before ordinal encoding
col_breed = ['Type', 'Breed1', 'Breed2']
df = data_test.X['cat'].copy()
for col in col_breed:
    df[col] = model_dict['cat_enc'][col].inverse_transform(df[col].values.reshape(-1,1))
# find unique (Type, Breed1, Breed2)
df_breed = df[['Type', 'Breed1', 'Breed2']].drop_duplicates().reset_index(drop=True)
breed_dict = df_breed.to_dict('index')
df_breed['Breed'] = df_breed.index
df = df.merge(df_breed, how='left', on=['Type', 'Breed1', 'Breed2'])
# get categorical label
df_label = pd.read_csv('breed_labels.csv')
breed_label = {breed['BreedID']: breed['BreedName'] for i, breed in df_label.iterrows()}
df['BreedName1'] = df['Breed1'].apply(lambda breed: breed_label.get(breed, None))
df['BreedName2'] = df['Breed2'].apply(lambda breed: breed_label.get(breed, None))
type_label = {1: 'Dog', 2: 'Cat'}
df['TypeName'] = df['Type'].apply(lambda x: type_label[x])
# plot 2D t-SNE 
k = 500
n = 30
emb_2s = emb2d[k:k+n]
plt.figure(figsize=(12, 12))
plt.scatter(emb_2s[:,0], emb_2s[:, 1], c=df['Breed'][k:k+n])
for i in range(n):
    row = df.loc[k+i]
    breed = row['Breed']
    animal_type = row['TypeName']
    breed1 = row['BreedName1']
    breed2 = row['BreedName1']
    plt.text(emb_2s[:,0][i], emb_2s[:,1][i], f'({animal_type}, {breed1}, {breed2})')
```

```python
# plot 3D t-SNE
n = 15
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')
emb_3s = emb3d[k:k+n]
ax.scatter(emb_3s[:,0], emb_3s[:,1], emb_3s[:,2], c=data_test.X['cat']['Type'][:n])
for i in range(n):
    row = df.loc[k+i]
    breed = row['Breed']
    animal_type = row['TypeName']
    breed1 = row['BreedName1']
    breed2 = row['BreedName1']
    ax.text(emb_3s[:,0][i], emb_3s[:,1][i], emb_3s[:, 2][i], f'({animal_type}, {breed1}, {breed2})')
```

The t-SNE visualizations showed that the embedding network was able to obtain somewhat meaningful representations of the categorical features; categorical features that have similar (Type, Breed1, Breed2) labels are generally clustered together. The embeddings were trained with 12 other categorical features, which could contribute to some entries data entries being further away from its usual (Type, Breed1, Breed2)-cluster.
